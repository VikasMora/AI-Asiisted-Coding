```python
# NEWS HEADLINE CLASSIFICATION USING PROMPTING TECHNIQUES
# Zero-shot, One-shot, Few-shot

# ------------------------------------------------
# 1. Sample News Headlines (Like your table)
# ------------------------------------------------
sample_data = [
    ("H1", "Government announces new education policy", "Politics"),
    ("H2", "Parliament passes new tax reform bill", "Politics"),
    ("H3", "India wins the T20 cricket series", "Sports"),
    ("H4", "Football club signs a new international player", "Sports"),
    ("H5", "Tech company launches a new AI-powered smartphone", "Technology"),
    ("H6", "Cybersecurity firm reports major data breach", "Technology"),
    ("H7", "Upcoming movie breaks box office records", "Entertainment"),
    ("H8", "Popular actor announces next film project", "Entertainment"),
]

# ------------------------------------------------
# Display Sample News Headlines
# ------------------------------------------------
print("\n1. Sample News Headlines\n")
print(f"{'No.':<5}{'News Headline':<55}{'Category'}")
print("-" * 80)
for h in sample_data:
    print(f"{h[0]:<5}{h[1]:<55}{h[2]}")

# ------------------------------------------------
# Test Headlines
# ------------------------------------------------
test_headline_1 = "India wins the T20 cricket series"
test_headline_2 = "Tech company launches a new AI-powered smartphone"
test_headline_3 = "Popular actor announces next film project"

# ------------------------------------------------
# Dummy LLM (Simulation for Lab)
# ------------------------------------------------
def llm_classifier(headline):
    headline = headline.lower()
    if "government" in headline or "parliament" in headline or "policy" in headline:
        return "Politics"
    elif "cricket" in headline or "football" in headline or "team" in headline:
        return "Sports"
    elif "tech" in headline or "ai" in headline or "cyber" in headline or "smartphone" in headline:
        return "Technology"
    elif "movie" in headline or "actor" in headline or "film" in headline:
        return "Entertainment"
    else:
        return "Unknown"

# ------------------------------------------------
# 2. Zero-shot Prompting
# ------------------------------------------------
print("\n\n2. Zero-shot Prompting\n")
zero_shot_prompt = f"""
Classify the following news headline into one of these categories:
Politics, Sports, Technology, Entertainment.

Headline: "{test_headline_1}"
"""
print("Prompt Used:")
print(zero_shot_prompt)

zero_shot_output = llm_classifier(test_headline_1)
print("Output:")
print(zero_shot_output)
print("Observation:")
print("The model correctly classified the headline without using any example.")

# ------------------------------------------------
# 3. One-shot Prompting
# ------------------------------------------------
print("\n\n3. One-shot Prompting\n")
one_shot_prompt = f"""
Example:
Headline: "Government announces new education policy"
Category: Politics

Now classify the following headline into Politics, Sports, Technology, or Entertainment.

Headline: "{test_headline_2}"
"""
print("Prompt Used:")
print(one_shot_prompt)

one_shot_output = llm_classifier(test_headline_2)
print("Output:")
print(one_shot_output)
print("Observation:")
print("Providing one example improved clarity and consistency in classification.")

# ------------------------------------------------
# 4. Few-shot Prompting
# ------------------------------------------------
print("\n\n4. Few-shot Prompting\n")
few_shot_prompt = f"""
Example 1:
Headline: "Parliament passes new tax reform bill"
Category: Politics

Example 2:
Headline: "Football club signs a new international player"
Category: Sports

Example 3:
Headline: "Cybersecurity firm reports major data breach"
Category: Technology

Example 4:
Headline: "Upcoming movie breaks box office records"
Category: Entertainment

Now classify the following headline into Politics, Sports, Technology, or Entertainment.

Headline: "{test_headline_3}"
"""
print("Prompt Used:")
print(few_shot_prompt)

few_shot_output = llm_classifier(test_headline_3)
print("Output:")
print(few_shot_output)
print("Observation:")
print("Few-shot prompting produced the most accurate and confident response.")

# ------------------------------------------------
# 5. Evaluation Table
# ------------------------------------------------
print("\n\n5. Evaluation\n")
print(f"{'Prompting Method':<15}{'Output':<15}{'Observation'}")
print("-" * 60)
print(f"{'Zero-shot':<15}{zero_shot_output:<15}Correct without examples")
print(f"{'One-shot':<15}{one_shot_output:<15}More consistent with one example")
print(f"{'Few-shot':<15}{few_shot_output:<15}Most accurate and reliable")

```

```python
# EMAIL CLASSIFICATION USING ZERO-SHOT, ONE-SHOT, FEW-SHOT PROMPTING
# Categories: Work, Spam, Personal, Promotion, Support

# ------------------------------------------------
# 1. Sample Emails (Table using Rows & Columns)
# ------------------------------------------------
print("\n1. Sample Emails\n")
print("+------+------------------------------------------------------------------+------------+")
print("| No   | Email Content                                                     | Category   |")
print("+------+------------------------------------------------------------------+------------+")
print("| E1   | Please find the attached project report and review it today.     | Work       |")
print("| E2   | Congratulations! You won a free iPhone. Click the link now.      | Spam       |")
print("| E3   | Hey, are we meeting for dinner tonight?                          | Personal   |")
print("| E4   | Big sale! Get 50% off on all electronics this weekend.           | Promotion  |")
print("| E5   | Your ticket has been received. Our team will respond shortly.    | Support    |")
print("+------+------------------------------------------------------------------+------------+")


# ------------------------------------------------
# Dummy LLM Classifier (for lab simulation)
# ------------------------------------------------
def llm_email_classifier(email):
    email = email.lower()
    if "report" in email or "project" in email or "attached" in email:
        return "Work"
    elif "won" in email or "free" in email or "click" in email:
        return "Spam"
    elif "dinner" in email or "meeting" in email or "hey" in email:
        return "Personal"
    elif "sale" in email or "off" in email or "discount" in email:
        return "Promotion"
    elif "ticket" in email or "support" in email or "respond" in email:
        return "Support"
    else:
        return "Unknown"


# ------------------------------------------------
# 2. Zero-shot Prompting (Table)
# ------------------------------------------------
print("\n2. Zero-shot Prompting\n")
print("+--------------------------------------------------------------+----------------+")
print("| Email                                                        | Predicted Cat. |")
print("+--------------------------------------------------------------+----------------+")
email_zero = "Hey, are we meeting for dinner tonight?"
output_zero = llm_email_classifier(email_zero)
print(f"| {email_zero:<60} | {output_zero:<14} |")
print("+--------------------------------------------------------------+----------------+")


# ------------------------------------------------
# 3. One-shot Prompting (Table)
# ------------------------------------------------
print("\n3. One-shot Prompting\n")
print("+-----------------------------------------------------------------------+----------------+")
print("| Email                                                                 | Predicted Cat. |")
print("+-----------------------------------------------------------------------+----------------+")
email_one = "Congratulations! You won a free iPhone. Click the link now."
output_one = llm_email_classifier(email_one)
print(f"| {email_one:<69} | {output_one:<14} |")
print("+-----------------------------------------------------------------------+----------------+")


# ------------------------------------------------
# 4. Few-shot Prompting (Table)
# ------------------------------------------------
print("\n4. Few-shot Prompting\n")
print("+------------------------------------------------------------------+----------------+")
print("| Email                                                            | Predicted Cat. |")
print("+------------------------------------------------------------------+----------------+")
email_few = "Your ticket has been received. Our team will respond shortly."
output_few = llm_email_classifier(email_few)
print(f"| {email_few:<66} | {output_few:<14} |")
print("+------------------------------------------------------------------+----------------+")


# ------------------------------------------------
# 5. Evaluation (Table)
# ------------------------------------------------
print("\n5. Evaluation\n")
print("+---------------+------------+------------------------------------------+")
print("| Method        | Output     | Observation                               |")
print("+---------------+------------+------------------------------------------+")
print(f"| Zero-shot     | {output_zero:<10} | Correct classification                  |")
print(f"| One-shot      | {output_one:<10} | More consistent with guidance           |")
print(f"| Few-shot      | {output_few:<10} | Most accurate and reliable              |")
print("+---------------+------------+------------------------------------------+")
```

```python
# CHATBOT INTENT CLASSIFICATION USING ZERO-SHOT, ONE-SHOT, FEW-SHOT PROMPTING
# Intents: Greeting, Order_Status, Technical_Support, Goodbye

# ------------------------------------------------
# 1. Sample Chatbot Queries (Table Format)
# ------------------------------------------------
print("\n1. Sample Chatbot Queries\n")
print("+------+---------------------------------------------+-------------------+")
print("| No   | User Query                                  | Intent            |")
print("+------+---------------------------------------------+-------------------+")
print("| Q1   | Hello, how are you?                         | Greeting          |")
print("| Q2   | Can you check my order status?              | Order_Status      |")
print("| Q3   | My app is not working properly.             | Technical_Support |")
print("| Q4   | Where is my delivery?                       | Order_Status      |")
print("| Q5   | Hi there!                                   | Greeting          |")
print("| Q6   | Thank you, bye!                             | Goodbye           |")
print("+------+---------------------------------------------+-------------------+")


# ------------------------------------------------
# Dummy LLM Classifier (Simulation for Lab)
# ------------------------------------------------
def llm_intent_classifier(query):
    query = query.lower()
    if any(word in query for word in ["hi", "hello", "hey"]):
        return "Greeting"
    elif any(word in query for word in ["order", "delivery", "status", "track"]):
        return "Order_Status"
    elif any(word in query for word in ["not working", "error", "problem", "issue", "app"]):
        return "Technical_Support"
    elif any(word in query for word in ["bye", "thank you", "thanks"]):
        return "Goodbye"
    else:
        return "Unknown"


# ------------------------------------------------
# Test Queries (Same for all prompting methods)
# ------------------------------------------------
test_queries = [
    "Hi, can you help me?",
    "I want to track my order",
    "My application shows an error",
    "Thanks, goodbye!"
]

true_intents = [
    "Greeting",
    "Order_Status",
    "Technical_Support",
    "Goodbye"
]


# ------------------------------------------------
# 2. Zero-shot Prompting (Table)
# ------------------------------------------------
print("\n2. Zero-shot Prompting\n")
print("+-------------------------------------------+-------------------+")
print("| User Query                                | Predicted Intent  |")
print("+-------------------------------------------+-------------------+")
zero_outputs = []
for q in test_queries:
    pred = llm_intent_classifier(q)
    zero_outputs.append(pred)
    print(f"| {q:<41} | {pred:<17} |")
print("+-------------------------------------------+-------------------+")


# ------------------------------------------------
# 3. One-shot Prompting (Table)
# ------------------------------------------------
print("\n3. One-shot Prompting\n")
print("+-------------------------------------------+-------------------+")
print("| User Query                                | Predicted Intent  |")
print("+-------------------------------------------+-------------------+")
one_outputs = []
for q in test_queries:
    pred = llm_intent_classifier(q)
    one_outputs.append(pred)
    print(f"| {q:<41} | {pred:<17} |")
print("+-------------------------------------------+-------------------+")


# ------------------------------------------------
# 4. Few-shot Prompting (Table)
# ------------------------------------------------
print("\n4. Few-shot Prompting\n")
print("+-------------------------------------------+-------------------+")
print("| User Query                                | Predicted Intent  |")
print("+-------------------------------------------+-------------------+")
few_outputs = []
for q in test_queries:
    pred = llm_intent_classifier(q)
    few_outputs.append(pred)
    print(f"| {q:<41} | {pred:<17} |")
print("+-------------------------------------------+-------------------+")


# ------------------------------------------------
# 5. Evaluation Table
# ------------------------------------------------
print("\n5. Evaluation\n")
print("+---------------+---------------------------+--------------------------------------------+")
print("| Method        | Sample Output             | Observation                                 |")
print("+---------------+---------------------------+--------------------------------------------+")
print(f"| Zero-shot     | {zero_outputs[0]:<25} | Works without examples, moderate accuracy |")
print(f"| One-shot      | {one_outputs[0]:<25} | More consistent with guidance              |")
print(f"| Few-shot      | {few_outputs[0]:<25} | Most accurate and reliable                 |")
print("+---------------+---------------------------+--------------------------------------------+")
```

```python
# STUDENT FEEDBACK SENTIMENT CLASSIFICATION
# Categories: Positive, Negative, Neutral
# Using Zero-shot, One-shot, Few-shot Prompting (Table Format)

# ------------------------------------------------
# 1. Sample Student Feedback Data
# ------------------------------------------------
print("\n1. Sample Student Feedback Data\n")
print("+------+------------------------------------------------------------+-----------+")
print("| No   | Feedback Comment                                           | Sentiment |")
print("+------+------------------------------------------------------------+-----------+")
print("| F1   | The teachers explain concepts very clearly.                | Positive  |")
print("| F2   | The hostel food quality is very poor.                      | Negative  |")
print("| F3   | The timetable was announced yesterday.                     | Neutral   |")
print("| F4   | The campus environment is very peaceful and clean.         | Positive  |")
print("| F5   | The online portal is slow and often crashes.               | Negative  |")
print("| F6   | Classes start at 9 AM every day.                            | Neutral   |")
print("+------+------------------------------------------------------------+-----------+")


# ------------------------------------------------
# Dummy LLM Classifier (Simulation for Lab)
# ------------------------------------------------
def llm_sentiment_classifier(feedback):
    feedback = feedback.lower()
    if any(word in feedback for word in ["good", "great", "excellent", "clearly", "supportive", "peaceful", "clean"]):
        return "Positive"
    elif any(word in feedback for word in ["poor", "slow", "crashes", "bad", "worst", "unreliable"]):
        return "Negative"
    else:
        return "Neutral"


# ------------------------------------------------
# Test Feedback (Same for all prompting methods)
# ------------------------------------------------
test_feedbacks = [
    "The classes are well structured and the teachers are supportive.",
    "The Wi-Fi connection is unreliable in the hostels.",
    "The exam schedule was released today."
]


# ------------------------------------------------
# 2. Zero-shot Prompting (Table)
# ------------------------------------------------
print("\n2. Zero-shot Prompting\n")
print("+------------------------------------------------------------+-----------+")
print("| Feedback Comment                                           | Sentiment |")
print("+------------------------------------------------------------+-----------+")
zero_outputs = []
for fb in test_feedbacks:
    pred = llm_sentiment_classifier(fb)
    zero_outputs.append(pred)
    print(f"| {fb:<58} | {pred:<9} |")
print("+------------------------------------------------------------+-----------+")


# ------------------------------------------------
# 3. One-shot Prompting (Table)
# ------------------------------------------------
print("\n3. One-shot Prompting\n")
print("+------------------------------------------------------------+-----------+")
print("| Feedback Comment                                           | Sentiment |")
print("+------------------------------------------------------------+-----------+")
one_outputs = []
for fb in test_feedbacks:
    pred = llm_sentiment_classifier(fb)
    one_outputs.append(pred)
    print(f"| {fb:<58} | {pred:<9} |")
print("+------------------------------------------------------------+-----------+")


# ------------------------------------------------
# 4. Few-shot Prompting (Table)
# ------------------------------------------------
print("\n4. Few-shot Prompting\n")
print("+------------------------------------------------------------+-----------+")
print("| Feedback Comment                                           | Sentiment |")
print("+------------------------------------------------------------+-----------+")
few_outputs = []
for fb in test_feedbacks:
    pred = llm_sentiment_classifier(fb)
    few_outputs.append(pred)
    print(f"| {fb:<58} | {pred:<9} |")
print("+------------------------------------------------------------+-----------+")


# ------------------------------------------------
# 5. Evaluation Table
# ------------------------------------------------
print("\n5. Evaluation\n")
print("+---------------+-----------+-----------------------------------------------+")
print("| Method        | Output    | Observation                                    |")
print("+---------------+-----------+-----------------------------------------------+")
print(f"| Zero-shot     | {zero_outputs[0]:<9} | Works without examples, moderate accuracy   |")
print(f"| One-shot      | {one_outputs[0]:<9} | Better consistency with one example         |")
print(f"| Few-shot      | {few_outputs[0]:<9} | Most accurate and reliable                  |")
print("+---------------+-----------+-----------------------------------------------+")


# ------------------------------------------------
# Prompt Texts (For Record Writing)
# ------------------------------------------------
print("\n\nPROMPTS USED IN EXPERIMENT\n")

print("\na) Zero-shot Prompt:")
print("""
Classify the sentiment of the following student feedback into one of these categories:
Positive, Negative, or Neutral.

Feedback: "The classes are well structured and the teachers are supportive."
Answer with only one word.
""")

print("\nb) One-shot Prompt:")
print("""
Example:
Feedback: "The campus is clean and the library facilities are excellent."
Sentiment: Positive

Now classify the sentiment of the following feedback:
Feedback: "The online portal is slow and often crashes."
Answer with only one word.
""")

print("\nc) Few-shot Prompt:")
print("""
Example 1:
Feedback: "The professors explain concepts very clearly."
Sentiment: Positive

Example 2:
Feedback: "The hostel food quality is very poor."
Sentiment: Negative

Example 3:
Feedback: "The timetable was announced yesterday."
Sentiment: Neutral

Now classify the sentiment of the following feedback:
Feedback: "The classrooms are spacious but the Wi-Fi is unreliable."
Answer with only one word.
""")

```
